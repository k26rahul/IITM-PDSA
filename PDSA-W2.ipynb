```python
def swap(arr, i, j):
  arr[i], arr[j] = arr[j], arr[i]
```

```python
def bubble_sort(arr):
  n = len(arr)

  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        swap(arr, j, j+1)

  return arr


bubble_sort([3, 2, 4, 5, 6, 7, 1, 0])
bubble_sort([1])
bubble_sort([])
```

```python
def selection_sort(arr):
  n = len(arr)

  for i in range(n):
    min_index = i

    for j in range(i+1, n):
      if arr[j] < arr[min_index]:
        min_index = j

    swap(arr, i, min_index)

  return arr


selection_sort([3, 2, 4, 5, 6, 7, 1, 0])
selection_sort([1])
selection_sort([])
```

```python
def insertion_sort(arr):
  n = len(arr)

  for i in range(n):
    j = i
    while (j > 0 and arr[j] < arr[j-1]):
      swap(arr, j, j-1)
      j = j-1

  return arr


insertion_sort([3, 2, 4, 5, 6, 7, 1, 0])
insertion_sort([1])
insertion_sort([])
```

```python
def merge(left, right):
  m, n = len(left), len(right)

  merged, i, j = [], 0, 0

  while i < m and j < m:
    if left[i] <= right[i]:
      merged.append(left[i])
      i += 1
    else:
      merged.append(right[j])
      j += 1
```

```python
def merge_sort(arr):
  n = len(arr)

  if n <= 1:
    return arr

  n_half = n//2

  left = arr[:n_half]
  right = arr[n_half:]
  return merge(left, right)


merge_sort([1, 0])
```
