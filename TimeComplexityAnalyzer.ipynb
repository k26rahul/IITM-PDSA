```python
from TimeComplexityAnalyzer import TimeComplexityAnalyzer
```

```python
# Linear Time Complexity (O(n))
def linear_time_function(n):
  for i in range(n):
    pass
```

```python
# Quadratic Time Complexity (O(n^2))
def quadratic_time_function(n):
  for i in range(n):
    for j in range(n):
      pass
```

```python
# Cubic Time Complexity (O(n^3))
def cubic_time_function(n):
  for i in range(n):
    for j in range(n):
      for k in range(n):
        pass
```

```python
# Exponential Time Complexity (O(2^n))
def exponential_time_function(n):
  if n <= 1:
    return n
  else:
    return exponential_time_function(n-1) + exponential_time_function(n-2)
```

```python
analyzer = TimeComplexityAnalyzer(linear_time_function)

for input_size in range(9):
  analyzer.run(10**input_size)

analyzer.plot()
```

```python
analyzer = TimeComplexityAnalyzer(quadratic_time_function)

for input_size in range(0, 10**3, 20):
  analyzer.run(input_size)

analyzer.plot()
```

```python
analyzer = TimeComplexityAnalyzer(cubic_time_function)

for input_size in range(0, 100):
  analyzer.run(input_size)

analyzer.plot()
```

```python
analyzer = TimeComplexityAnalyzer(exponential_time_function)

for input_size in range(1, 20):
  analyzer.run(input_size)

analyzer.plot()
```
